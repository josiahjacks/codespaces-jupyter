```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('CCA_Scores.csv')
```

```python
type (df)
```

```python
len (df)
```

```python
df.shape
```

```python
plt.figure(figsize=(8, 5))
plt.bar(df.index, df['Scores'])
plt.xlabel('Attempt')
plt.ylabel('Scores')
plt.title('Bar Graph of Scores')
plt.show()
```

```python
plt.figure(figsize=(8, 8))
plt.pie(df['Scores'], labels=df.index, autopct='%1.1f%%', startangle=90)
plt.title('Pie Chart of Scores')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(df.index, df['Scores'])
plt.xlabel('Attempt')
plt.ylabel('Scores')
plt.title('Scatter Plot of Scores')
plt.show()
```
